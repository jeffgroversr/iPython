## Scikit-Learn 4-Step Modeling Patterns

- Step 1: Import the class
- Step 2: Instantiate the Estimator
    a. Estimator is a scikit-learn's term for model
    b. Instantiate means make an instance of
- Step 3: Fit the model with data
    a. Model learns relationships between features and responses (X and y)
    b. Occurs in place

## Step 1

```python
#Step 1
from sklearn.neighbors import KNeighborsClassifier
```

```python
# import Load_iris function from datasets module
from sklearn.datasets import load_iris
```

```python
# sace "bunch" object containing iris dataset and its attributes
iris = load_iris()
type(iris)
```

```python
# print the iris data
iris.data
```

```python
# print the names of the four features
iris.feature_names
```

```python
# print integers representing the species of each object
iris.target #target represents what we are going to predict
```

```python
# print the encoding scheme for species: 0 = setosa, 1 = versicolor, 2 = virginica
iris.target_names
```

```python
# check the types of the features and response
type(iris.data)
```

```python
type(iris.target)
```

```python
# check the shape of the features (first dimension = # of observations, secon dimensions = number of features)
iris.data.shape
```

```python
# check the shape of the response (single dimension matching the number of observations)
iris.target.shape
```

```python
# store feature matrix in "X"
X = iris.data
```

```python
# store feature matrix in "y"
y = iris.target
```

## Step 2

```python
#Step 2
knn = KNeighborsClassifier(n_neighbors=1) #Knn object is looking for 1 nearest neighbor
knn
```

## Step 3

```python
#Step 3
knn.fit(X,y)
```

## Step 4

```python
knn.predict([3,5,4,2])
```

```python
X_new = [[3,5,4,2],[5,4,3,2]]
knn.predict(X_new) #Prediction for the first unknown iris = 2, and the first = 1
```

```python
# Model tuning
# instantiate the model (using K = 5)
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#fit the model with data
knn.fit(X,y)
```

```python
# Predict the response for new observations
knn.predict(X_new)
```

## Using a different classification model

```python
# Step 1: Import the class
from sklearn.linear_model import LogisticRegression
```

```python
# Step 2: instantiate the model with default parameters
logreg = LogisticRegression()
```

```python
# Step 3: fit the model with data
logreg.fit(X,y)
```

```python
# Step 4: predict the response for new observations
logreg.predict(X_new) #Model predicts a value of 2 for the first unknown iris and 0 for the second
```
