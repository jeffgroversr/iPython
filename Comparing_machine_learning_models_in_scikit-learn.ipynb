## Evaluation procdure # 1: train and test on the entire dataset
- 1. Train the model on the entire dataset
- 2. Test the model on the same dataset and evaluate how well we did by comparing th predicted responses values with the true response values

```python
# read in the iris data
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#create x (features) and y(response)
X = iris.data
y = iris.target
```

## Logistic regression

```python
# Step 1: Import the class
from sklearn.linear_model import LogisticRegression
```

```python
# Step 2: instantiate the model with default parameters
logreg = LogisticRegression()
```

```python
# Step 3: fit the model with data
logreg.fit(X,y)
```

```python
# Step 4: predict the response for new observations
logreg.predict(X) #Model predicts a value of 2 for the first unknown iris and 0 for the second
```

```python
#store the predicted response values
y_pred = logreg.predict(X)
```

```python
#check how many predictions were generated
len(y_pred)
```

##Classification accuracy

```python
#Compute classification accuracy for the Logistic regression model
from sklearn import metrics
metrics.accuracy_score(y, y_pred) #training accuracy
```

## Evaluation procedure # 2: Train/test split
- 1. Split the data into training and testing
- 2. Train the model on the training set
- 3. Test the model

```python
# print the shapres of X and y
X.shape
```

```python
y.shape
```

```python
# Step 1: Split X and y into training and testing sets
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4)
```

```python
# print the shapres of the X objectives
X_train.shape
```

```python
# print the shapres of the X objectives
X_test.shape
```

```python
# print the shapres of the X objectives
X_test.shape, X_train.shape
```

```python
# print the shapres of the y objects
y_train.shape
```

```python
y_test.shape
```

```python
# print the shapres of the X objectives
y_test.shape, y_train.shape
```

```python
#Step 2: train the model on the training set
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
#Step 3: make predictions on the testing set
y_pred = logreg.predict(X_test)
```

```python
# compare actual response values (y_test) with predicted response values (y_pred)
metrics.accuracy_score(y_test, y_pred)
```

Repear for K=5

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

Repeat for K=1

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

Can we locate an even better K?

```python
#Try K=1 through K=25 and record testing accuracy
k_range = range(1, 26)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
```

```python
#import Matplotlib
import matplotlib.pyplot as plt
#allow plots to appear within the notebook
%matplotlib inline

#plot the relationship between K and testing accuracy
plt.plot(k_range, scores)
plt.xlabel('Value for K for KNN')
plt.ylabel('Testing Accuracy')
```
